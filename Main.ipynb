```python
import os
from langchain_community.llms import HuggingFaceHub
from pypdf import PdfReader 
from pprint import pprint
import os 

HF_token = os.getenv("HUGGING_FACE_TOKEN")

os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_ENDPOINT"] = "https://api.smith.langchain.com"
os.environ["LANGCHAIN_PROJECT"] = "Portfolio Gen"
os.environ["LANGCHAIN_API_KEY"] = os.getenv("LANGCHAIN_API_KEY")

text = ""
linkedin_prof = PdfReader("LinkedIn profile.pdf")

for page in linkedin_prof.pages:
    text += page.extract_text()
pprint(text)
```

```python
from langchain_huggingface import HuggingFaceEndpoint

llm = HuggingFaceEndpoint(
    repo_id="bigcode/starcoder2-3b",
    task="text-generation",
    max_new_tokens=2500,
    do_sample=False,
    huggingfacehub_api_token=HF_token
)

prompt = f"Write an HTML code to create a portfolio webpage based on {text}."
output = llm.invoke(prompt)
```

```python
print(output)
with open("portfolio.html", "w", encoding="utf-8") as f:
    f.write(output)
```
